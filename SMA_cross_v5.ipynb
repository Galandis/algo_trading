```python
import IterativeBacktest as IB
import pandas as pd
import numpy as np
import json
```

```python
# Import prepared dictionary which contains average spreads for all assets
with open('Data\spreads.json', 'r') as fp:
    spreads = json.load(fp)

# Prepare instruments list from spreads file
instruments_list = []
for key in spreads:
    instruments_list.append(key)
instruments_list = instruments_list[1:]
```

```python
timeframe = "M5"
amount = 10000
smas = range(20, 50, 2)
smal = range(100, 200, 5)
```

```python
i = 0
j = len(instruments_list)
```

```python
#testing single symbols
instruments_list = ['SOYBN_USD']
smas = [48]
smal = [185]
```

```python
for ticker in instruments_list:
    i += 1
    # Prepare dataframe to save all outcomes
    test_outcome = pd.DataFrame(columns = ['sma_s','sma_l','trades','performance'])
    spread = spreads[ticker]
    bc = IB.IterativeBacktest(ticker, timeframe, spread, amount, use_spread=False)
    for sma_s in smas:
        # Prepare dataframe to save all outcomes
        for sma_l in smal:
            outcome = bc.test_sma_strategy(sma_s,sma_l)
            test_outcome = test_outcome.append(outcome)
        # Saving to CSV if required
        #test_outcome.to_csv('Wyniki\SMA\{}_{}.csv'.format(ticker,timeframe),index=True)
    print("Part {}/{} is done".format(i,j))
```

## Checking best values of SMAs for each symbol

```python
summarize = pd.DataFrame(columns = ['symbol','spread','sma_s','sma_l','trades','Perf(%)','OverallLosing(%)'])
```

```python
for ticker in instruments_list:
    spread = spreads[ticker]
    data = pd.read_csv('Wyniki\SMA\{}_{}.csv'.format(ticker,timeframe),index_col=['Unnamed: 0'])
    #Looking for best performance
    best = data.sort_values(by = ['performance'], ascending=False).head(1)
    best = best.squeeze() #to remove single-dimensional entries from the shape of an array
    # Counting how many strategies are below 0
    minus = data.where(data.performance < 0)
    minus.dropna(inplace=True)
    losing = len(minus)/len(data)*100
    wynik = pd.DataFrame([[ticker,spread,best.sma_s,best.sma_l,best.trades,best.performance,losing]],
                        columns=['symbol','spread','sma_s','sma_l','trades','Perf(%)','OverallLosing(%)'])
    summarize = summarize.append(wynik)
```

```python
summarize.set_index('symbol', inplace=True)
summarize
```

```python
summarize.to_csv('Wyniki\SMA\WynikiSMA.csv', index=True)
```

### After having a look into the summarize table there are two things which needs to be clarified

#### First: It is doubtful that after making trades the performance of strategy will be equal 0%

```python
zeros = summarize.where(summarize['Perf(%)'] == 0)
zeros.dropna(inplace=True)
zeros
```

```python
zeros.to_csv('Wyniki\SMA\BrakiSMA.csv', index=True)
```

#### Price of the asset is greater than invested amount. With SMA one should not use leverage to enter the position hence no position should be made. The assets needs to be dropped.

```python
summarize = summarize.where(summarize['Perf(%)'] != 0)
summarize.dropna(inplace=True)
summarize
```

```python
summarize.to_csv('Wyniki\SMA\WynikiSMA.csv', index=True)
```

#### Second: The OverallLosing(%) columns indicates that all pairs of SMAs: SMAs in range 20 - 50 - 2 and SMAl in range 100 - 200 - 5 had a profitable final outcome.
##### For below seven symbols SMA strategy is profitable in all the range.

```python
profitables = summarize.where(summarize['OverallLosing(%)'] == 0)
profitables.dropna(inplace=True)
profitables
```

```python
profitables.to_csv('Wyniki\SMA\ProfitSMA.csv', index=True)
```
